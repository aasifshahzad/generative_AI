```python
from openai import OpenAI
from dotenv import load_dotenv, find_dotenv

_ : bool = load_dotenv("E:/Python/.env")
client: OpenAI = OpenAI()
```

```python
# Arabic Audio Input

audio_file = open("arabic.mp3", "rb")
transcript_arabic = client.audio.transcriptions.create(
  model="whisper-1", 
  file=audio_file, 
  response_format="text"
)
transcript_arabic
```

```python
# urdu_arabic Audio Input

audio_file = open("urdu_arabic.mp3", "rb")
transcript_urdu_arabic = client.audio.transcriptions.create(
  model="whisper-1", 
  file=audio_file, 
  response_format="text"
)
transcript_urdu_arabic
```

```python
#Translation
# urdu_arabic Audio Input

audio_file = open("urdu_arabic.mp3", "rb")
transcript_urdu_arabic = client.audio.translations.create(
  model="whisper-1", 
  file=audio_file, 
  response_format="text"
)
transcript_urdu_arabic
```
